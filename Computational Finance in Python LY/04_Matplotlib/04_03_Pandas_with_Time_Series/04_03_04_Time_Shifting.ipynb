# Time Shifting

```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('time_data/walmart_stock.csv',index_col='Date')
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

```python
df.tail()
```

## .shift() forward

```python
df.shift(1).head() # values are shifted down by one row
```

```python
# You will lose that last piece of data that no longer has an index!
df.shift(1).tail()
```

## shift() backwards

```python
df.shift(-1).head()
```

```python
df.shift(-1).tail()
```

```python
(df-df.shift(1))/df.shift(1)
# compute return easily
```
